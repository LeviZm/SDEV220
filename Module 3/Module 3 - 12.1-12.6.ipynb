```python
# 12.5

plain = {'a': 1, 'b': 2, 'c': 3}

print(plain)
```

```python
# 12.6

from collections import OrderedDict

fancy = OrderedDict(a=1, b=2, c=3)

print(fancy)
```

Yes, they printed in the same order as plain. This is because starting with Python 3.7, dictionaries retain keys in the order in which they are added.
