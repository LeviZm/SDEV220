```python
# 10.2

def get_odds():
    for number in range(10):
        if number % 2 == 1:
            yield number


count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print(odd)

```

```python
# 10.3

def test(func):
    def wrapper():
        print('start')
        result = func()
        print('end')
        return result
    return wrapper

@test
def print_name():
    print(good())

print_name()
```
