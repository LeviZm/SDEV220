```python
# Module 2 - 6.1

import random # import the random module to generate random numbers

secret_number = random.randint(1, 10)
guess = input("Guess a number between 1 and 10: ")
guess = int(guess)  # Convert input to integer for comparison

if guess > secret_number:
    print("too high")
elif guess < secret_number:
    print("too low")
else:
    print("just right")
```

We can also add a while statement to allow for continued guesses until we get the right number

```python
# Module 2 - 6.1 extra

import random # import the random module to generate random numbers

secret_number = random.randint(1, 10)
guess = input("Guess a number between 1 and 10: ")
guess = int(guess)  # Convert input to integer for comparison

while guess != secret_number:
    if guess > secret_number:
        print("too high")
        guess = input("Guess again: ")
        guess = int(guess)
    elif guess < secret_number:
        print("too low")
        guess = input("Guess again: ")
        guess = int(guess)
else:
    print("just right")
```
