```python
# Module 2 - 7.1

for x in range(3,-1,-1):
    print(f'{x}')
```

I can also use the reversed() function on a standard range to get the same effect

```python
for x in reversed(range(4)):
    print(f'{x}')
```

Or, I can use a list of the numbers to print them

```python
values = [3, 2, 1, 0]

for x in values:
    print(f'{x}')
```
