```python
# Module 2 - 6.2

small = True
green = True

items = ['cherry','pea','watermelon','pumpkin']

for x in items:
    if small and not green:
        print(f'{x}')
    elif small and green:
        print(f"{x}")
    elif not small and green:
        print(f"{x}")
    else:
        print(f"{x}")
```
